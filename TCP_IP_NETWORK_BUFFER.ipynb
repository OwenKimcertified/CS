### TCP

"TCP는 인터넷을 통해 데이터를 안정적으로 전송하기 위한 프로토콜" 이다.

통신하기 전에 먼저 연결을 설정하고, 데이터를 주고받은 후에는 연결을 해제함.

`NDB 에서 TCP 가 필요한 이유`

1. 하나의 네트워크로 묶인 개별의 data node 컨테이너들에 대하여

   네트워크 통신을 하고 데이터를 주고받기 위함이다.

    TCP 는 IP 와 같이 사용되고 

    `보내는` 쪽에서는 데이터를 패킷단위로 나누고 전송.

    `받는` 쪽에서는 패킷단위의 데이터를 재조립함.

2. NDB 클러스터에 접속하는 클라이언트는 SQL 노드를 통해 클러스터와 통신

### IP 

IP는 컴퓨터 네트워크에서 데이터를 주고받을 때 사용되는 주소 지정 시스템.

IP는 데이터를 패킷으로 나누어 목적지까지 전달하는 역할.

IP 패킷은 출발지 IP 주소와 목적지 IP 주소를 포함.

IP는 패킷을 전달하는 데 중점을 두고 있으며, 데이터의 손실이나 손상을 처리하지 않는다.

컴퓨터 간의 통신을 위한 주소 지정 및 패킷 라우팅

`NDB 에서 IP 가 필요한 이유`

데이터 노드 간 데이터를 주고 받기 위해서 데이터를 패킷 단위로 보내고 재조립하는 과정이 필요.

-----------------------------------

`NDB 에서 TCP / IP 가 필요한 이유 정리`

데이터 노드 컨테이너 두 개가 있을 때, 각각의 IP 가 있을 것이고, 

그 두 개를 연결한 네트워크에 두 개의 컨테이너가 존재할텐데,

인터넷(IP) 를 통해 두 데이터 노드간 데이터의 송/수신을 안정적으로 하기 위해서 

TCP 를 설정. 데이터를 주고받기 위한 연결을 설정. 데이터를 주고받고 난 후에는 TCP 를 해제하는 것이고

IP 는 TCP에 의해 통신을 위해 연결된 상태에서 데이터 노드의 주소(인터넷 프로토콜/IP) 를 가지고 

데이터를 패킷 단위로 보내고 받는 쪽에서는 패킷 단위의 데이터를 재 조립하여 목적지까지 데이터를 전달하는 것이다 

그래서 위와 같은 이유로 NDB Cluster 에서는 TCP 와 IP 가 필요함.

----------------------------------

`NDB Cluster config.ini` 에 있는 설정 

```python
# TCP PARAMETERS

[tcp default]
SendBufferMemory=2M
ReceiveBufferMemory=2M
```

위의 설정은 TCP 소켓에 대한 기본 TCP 매개변수를 정의하는 부분

이제는 알 수 있 듯, IP 로 데이터를 패킷 단위로 보낼 때 

`보내는 양` 과 `받아서 재조립 하는 양` 이다.

### Socket

트워크 상에서 데이터 통신을 위한 인터페이스를 제공하는 소프트웨어.

Server Socket 과 Client Socket 으로 데이터를 주고 받을 통로를 연결하는 부분.
